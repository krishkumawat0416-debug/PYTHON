```python
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop
 (columns=["total_bill"]), df["total_bill"], test_size=0.20)
```

```python
x_train
```

```python
oe = OrdinalEncoder(categories=[["Thur", "Fri", "Sat", "Sun"]])
x_train_day = oe.fit_transform(x_train[["day"]])
x_test_day = oe.transform(x_test[["day"]])
```

```python
x_train_day.shape
```

```python
ohe = OneHotEncoder(drop="first" , sparse_output =False)
x_train_smoker_day = ohe.fit_transform(x_train[["smoker" , "day"]])

x_test_smoker_day = ohe.transform(x_test[["smoker" , "day"]])
```

```python
x_train_smoker_day.shape
```

```python
x_train_tip = x_train.drop(columns=["smoker" , "day" , "sex" , "time", "size"])
x_test_tip = x_test.drop(columns=["smoker" , "day" , "sex" , "time" , "size"])
```

```python
x_train_transformed = np.concatenate((x_train_tip,
                                      x_train_smoker_day ,
                                      x_train_day) , axis=1)
```
