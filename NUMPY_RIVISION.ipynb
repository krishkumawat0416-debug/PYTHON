```python
# Arange --> last value exclude
arr = np.arange(1,11)
# here we are creating the view of the original array
view_arr = arr.view()
# here we are changing the first elements of the duplicated array

# view_arr[0] =  100
# view_arr

print(f'Original array : {arr}')
print(f'View : {view_arr}')

# View is a method which copies the refrence of the original object or
# arr . whenever we want to change anything into the duplicate array
# then it would effect on the original array
```

```python
copy_arr = arr.copy()
print(copy_arr)

copy_arr[1] = 20
print(copy_arr)
print(arr)
```
