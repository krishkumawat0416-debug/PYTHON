```python
import numpy as np
```

```python
a = np.random.randint(100, 200, 64).reshape(4,16)
a
```

```python
a = np.full((5,5),7)
a
```

```python
# view --> view is create a array you can change array element in use slicing then change array
# but change the original array in view

a = np.array([12,23,34,56,7,8,96,34,56])
b = a[5:9]
b[:] = 0
b
```

```python
# copy --> copy is create a array you can change array element in use slicing and change the element but not change
# the original array only change copy array
a = np.array([12,23,34,56,7,8,96,34,56])
b = a[5:9].copy()
b[:] = 0
b
```

```python
# linesapce --> is create a array in same gap of values
a = np.linspace(1,10,5)
a
```

```python
# unique --> in create a array and remove duplicate and return 3 values
# arr --> reurn unique elememt
# return_index = true --> return indexing
# return_count = true --> return count

a = np.array([1,2,8,4,5,3,3,3,3,3,1,1,1,2,2,5,6,4,7,8])
a = (np.unique(a, return_index = True, return_counts = True))
a
```

```python
# what is hstack() and vstack() ? create 4 arrays and connect it via  hstack, vstack. ?
# hstack --> it will add 2 or more array

a = np.arange(1,5).reshape(2,2)
a
```

```python
b = np.arange(5,9).reshape(2,2)
b
```

```python
c = np.arange(9,13).reshape(2,2)
c
```

```python
d = np.arange(13,17).reshape(2,2)
d
```

```python
np.hstack((a,b,c,d))
```

```python
# vstack --> it will Add 2 or more array
a = np.arange(1,5).reshape(2,2)
b = np.arange(5,9).reshape(2,2)
c = np.arange(9,13).reshape(2,2)
d = np.arange(13,17).reshape(2,2)

np.vstack((a,b,c,d))
```

```python
# create an array in range 11-26 now you have to reshape it (4*4) and then you have to
# calculate total sum, row wise sum,  column wise sum

a = np.arange(11,27).reshape(4,4)
print(a.sum(axis = 0))
```

```python
a = np.arange(11,27).reshape(4*4)
print(a.sum(axis = 1))
```
