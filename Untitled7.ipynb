```python
data = {'a': 10, 'b': 5, 'c': 20, 'd': 3}

keys = list(data.keys())
min = keys[0]
value = data[min]

for key in keys[1:]:
    if data[key] < value:
        value = data[key]
        min = key

print(min, ":", value)
```
