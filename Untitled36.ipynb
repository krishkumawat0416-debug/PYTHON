```python
# pandas have two different data types
# (1). Series
# (2). Data Frame
```

```python
# Series create with the help of numpy array
s = pd.Series(np.array([1,2,3,4]), index=['a','b','c','d'])
```

```python
s.size
```

```python
# Creating a series with the help of list
li = [1,2,3,4]
li_idx = ['a', 'b', 'c', 'd']
pd.Series(li, index=li_idx)
```

```python
# Creating a series with the help of dictionary
s1 = d = {'a':10,
     'b':20,
     'c':30,
     'd':40}
# When we create any series with the help of dictionary
# key --> index , value --> value
s1 = pd.Series(d)
```

```python
# values is used to retrieve the values of the series
# it returns the values into the form of the numpy array
s1.values
```

```python
# index method is used to find all the indices of the series
s1.index
```

```python
s1.dtype
```

```python
# Vectorised oprations
s1 = pd.Series([1,2,3,4])
s2 = pd.Series([5,6,7,8])

s1 + s2
```

```python
s1 ** 2
```

```python
# Accessing the elements using indices
series = pd.Series(np.arange(1,13))
# We can access the elements of the series using some conditions
series[series > 5]
```

```python
series[series % 2 == 0]
```

```python
# Data Frame --> row and columns

# Creating the data frame with help of dictionary

student_details = {
    'Name' : ['krish' , 'rahul', 'aman'],
    'Subject' : ['Maths', 'English', 'Science'],
    'Marks' : [80, 90, 95],
    'Address' : ['Jaipur' , 'Delhi' , 'Mp']
}

pd.DataFrame(student_details)
```

```python
student_detail = [
    ['krish', 'Maths', 80, 'Jaipur'],
    ['rahul', 'English', 90, 'Delhi'],
    ['aman', 'Science', 95, 'Mp']
]

pd.DataFrame(student_detail, columns=['Name', 'Subject', 'Marks', 'Address'])
```
