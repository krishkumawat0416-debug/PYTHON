```python
# Task ::: We have to create a dataframe of employees in which column are
# emo_id, emp_name, department, salary, working_hours
```

```python
a = {
    "emp_id": ['A101', "A102", 'A103'],
    "emp_name":['krish', 'rehan', 'abhishek'],
    "department":['it', 'it', 'op'],
    "salary":[20000, 23000, 45000],
    "working_hours":[6, 8, 10]
}
```

```python
df = pd.DataFrame(a)
```

```python
df
```

```python
# How can we export our dataframe into csv data ?
```

```python
df.to_csv("emp_info.csv")  ### to_csv means export your data into csv file
```

```python
df.to_csv("new_emp_info.csv", index = False)  ## index = false means it willl remove unnamed column
```
