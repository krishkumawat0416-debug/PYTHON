```python
# using function
def rev(start, end, li):
    while start < end:
          li[start], li[end] = li[end], li[start]
          start += 1
          end -= 1

li = [1,2,3,4,5]
k = 2

rev(0, len(li) - 1, li)
rev(0, k - 1, li)
rev(k, len(li) - 1, li)
print(li)
```

```python
# bank management stsyem
# global vatiable

balance = 1000

# function for the withraw amount
def withdraw(amount):
    global balance
    if amount > 0 and balance >= amount:
       balance -= amount
       print(f"you've suessfully withdraw {amount}")
       print(f"Your balance is {balance}")
    else:
        print("enter valid amount")

# function for deposit the balance

def deposit(amount):
    global balance
    if amount > 0:
       balance += amount
       print(f"you've suessfully deposit {amount}")
       print(f"Your remaining balance is {balance}")


deposit(100)
withdraw(200)
```

```python
# student grading sydtem
# nested functions

def grade(marks):
   #nested function
   def calculate_percentage(marks):
      p = (marks/100)*100
      return p
   # nested function
   def assign_grade(percentage):
      if percentage > 90 and percentage <= 100:
         return "A"
      elif percentage > 80 and percentage <= 90:
         return "B"
      elif percentage > 70 and percentage <= 80:
          return "C"
      else:
          return "fail"
   # here we're calling the calculate percentange function (marks -> percentage)
   percent = calculate_percentage(marks)

   # here we're calling the assign grade function(percentage -> grade)
   grad = assign_grade(percent)

   # marks -> grade
   return grad

Grade = grade(91)
print(Grade)
```

```python
# writing the n number  fibbonacci number

def fibb(n):
   # base case

     if n == 1:
        return 1
     if n == 0:
        return 0
     return fibb(n-1) + fibb(n-2)
#  print(fibb(3))

total_sum = 0
for i in range(10):
     fib_num = fibb(i)
     print(fib_num, end = " ")
     total_sum += fib_num
print("\nSum of the first 10 Fibonacci numbers:", total_sum)
```

```python
# print factorial

def fact(n):
   if n == 1 or n == 0:
      return 1
   return n * fact(n-1)
print(fact(5))
```

```python
# factorial
n = 5
fact = 1
for i in range(1, n+1):
   fact *= i
print(fact)
```

```python
def reverse(left, right, li):
    while left < right:
        li[left], li[right] = li[right], li[left]
        left += 1
        right -= 1

li = [1,2,3,4,5]
k = 3
reverse(0, len(li) - 1, li)
reverse(0, k - 1, li)
reverse(k, len(li) - 1, li)

print(li)
```

```python
class car:
    def __init__(self, wheels, colour, engine, bulletproof, sunroof):
        self.wheels = wheels
        self.colour = colour
        self.engine = engine
        self.bulletproof = bulletproof
        self.sunroof = sunroof
    def display_info(self):
        print(f"wheels: {self.wheels}")
        print(f"colour: {self.colour}")
        print(f"engine: {self.engine}")
        print(f"bulletproof: {self.bulletproof}")
        print(f"sunroof: {self.sunroof}")

car1 = car("mrf", "black", "v8", True, True)

print(car1.bulletproof)
```
